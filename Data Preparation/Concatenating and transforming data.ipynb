# Data Preparation Basics
## Segment 4 - Concatenating and transforming data

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
DF_obj = pd.DataFrame(np.arange(100).reshape(10,10))
DF_obj
```

```python
DF_obj_2 = pd.DataFrame(np.arange(25).reshape(5,5))
DF_obj_2
```

### Concatenating data

```python
pd.concat([DF_obj, DF_obj_2])
```

```python
pd.concat([DF_obj, DF_obj_2], axis=1)
```

### Transforming data
#### Dropping data

```python
DF_obj.drop([1,5])
```

```python
DF_obj.drop([2,5,8], axis=1)
```

### Adding data

```python
series_obj = Series(np.arange(9))
series_obj.name = "added_variable"
series_obj
```

```python
variable_added = DataFrame.join(DF_obj, series_obj)
variable_added
```

```python
added_datatable = variable_added.append(variable_added, ignore_index=False)
added_datatable
```
